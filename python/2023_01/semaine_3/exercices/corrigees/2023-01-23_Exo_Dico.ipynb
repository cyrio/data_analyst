## Compter les mots 
### Remplacer tous les signes de ponctuation par des espaces

```python
punctuation = set(".;,?!:-_\"'\n\t’"+ string.punctuation)
for p in punctuation:
    data = data.replace(p," ")
```

## Mettre le texte en minuscules

```python
data = data.lower()
```

## Eliminer les blancs multiples 

```python
while '  ' in data:
    data = data.replace('  ', ' ')
```
