```python
def contient(words,pattern):
    res = False
    for mot in words:
        if pattern in mot:
            if len(mot) != 0 and len(pattern) == 0:
                res = False
            else:
                res = True
    return res


print(contient(["riri","fifi","loulou"],"i"))   
print(contient(["riri","fifi","loulou"],""))
print(contient(["riri","fifi","loulou",""],""))
```

```python
from math import ceil, sqrt
def Diviseur(n):
    ens_div = set()
    for item in range(1,ceil(sqrt(n)+1)):
        if n%item == 0:
            ens_div |= {item, n%item}
        ens_div -= {0,1}
    return ens_div

print(Diviseur(100))
```

```python
def est_premier(n):
    return len(Diviseur(n)) == 1

def ListedesPremiers(N):
    liste_premier = []
    for i in range(N+1):
        if est_premier(i):
            liste_premier.append(i)
    return liste_premier
            
print(ListedesPremiers(100))
```
