    Python est un langage interprété. Cela signifie qu’il n’est pas analysé et traduit en langage machine une fois pour toutes mais à chaque exécution du programme et qu’il est nécessaire d’utiliser un interpréteur à chaque exécution.

- **Que signifie REPL ?**
        
        
    REPL signifie Read Evaluate Print Loop, et est le nom donné à l'invite MicroPython interactive accessible sur les appareils Pycom. L'utilisation du REPL est de loin le moyen le plus simple de tester le code Python et d'exécuter des commandes. Vous pouvez utiliser le REPL en plus d'écrire des scripts dans main.py .
        
        
- **Qu’est ce qu’un IDE ?**
        
        
    Un environnement de développement intégré (IDE) est un logiciel permettant de créer des applications qui combinent des outils de développement courants dans une seule interface utilisateur graphique (GUI). Un IDE se compose généralement de :

    Éditeur de code source : un éditeur de texte qui peut aider à écrire du code logiciel avec des fonctionnalités telles que la coloration syntaxique avec des repères visuels, l'auto-complétion spécifique à la langue et la recherche de bogues au fur et à mesure de l'écriture du code.
    
    Automatisation de la construction locale : utilitaires qui automatisent des tâches simples et reproductibles dans le cadre de la création d'une version locale du logiciel à utiliser par le développeur, comme la compilation du code source de l'ordinateur en code binaire, l'empaquetage du code binaire et l'exécution de tests automatisés.
    
    Débogueur : un programme pour tester d'autres programmes qui peuvent afficher graphiquement l'emplacement d'un bogue dans le code d'origine.

- **Qu’est ce que le site StackOverflow ?**

    Stack Overflow est un site Web de questions et réponses pour les programmeurs professionnels et passionnés. Le site Web sert de plate-forme aux utilisateurs pour poser et répondre à des questions et, grâce à l'adhésion et à la participation active, pour voter les questions et les réponses vers le haut ou vers le bas comme Reddit et modifier les questions et les réponses d'une manière similaire à un wiki. Les utilisateurs de Stack Overflow peuvent gagner des points de réputation et des "badges" ; par exemple, une personne reçoit 10 points de réputation pour avoir reçu un vote « positif » sur une question ou une réponse à une question, et peut recevoir des badges pour ses précieuses contributions, ce qui représente une gamification du site Web traditionnel de questions-réponses. Les utilisateurs débloquent de nouveaux privilèges avec une augmentation de la réputation, comme la possibilité de voter, de commenter et même de modifier les publications d'autres personnes.
