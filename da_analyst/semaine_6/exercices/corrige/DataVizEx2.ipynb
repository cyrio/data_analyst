```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
%%time
df = pd.read_csv("./data/sales_predictions.csv")
print('dataset loaded')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.relplot(x='date', y='item_price', data=df, kind='line')
plt.show()
```

```python
sns.relplot(x="date", y="item_price", data = df.sample(100), kind="line", height=8, aspect=3)
```

```python
df.date = pd.to_datetime(df.date)
print('date ok')
```

```python
f, ax = plt.subplots(figsize=(20,6))
sns.relplot(x="date", y="item_price", data = df, kind="line", ax = ax)
plt.close(2)
```
