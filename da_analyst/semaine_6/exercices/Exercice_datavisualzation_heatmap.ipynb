```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_house = pd.read_csv("../cours/csv_sample/house_pricing.csv")
```

```python
df_house_map = df_house.iloc[:,-15:]
```
